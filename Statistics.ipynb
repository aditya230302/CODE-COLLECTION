# **Sampling methods**
- Sampling methods are ways to select a sample of data from a given population (every individual in the whole group).

It is unrealistic to collect data from the entire population because it:
- is too big
- takes too much time
- costs too much money
We therefore take an appropriate sized sample as a way of representing the population.

---

### **Types of Sampling Methods**

| **Sampling Method**       | **Description**                                                                                                                                         | **Example**                                                                                                                                                         |
|---------------------------|---------------------------------------------------------------------------------------------------------------------------------------------------------|---------------------------------------------------------------------------------------------------------------------------------------------------------------------|
| **Random Sampling**       | • Gathering a representative sample from a population where each member in the population has an **equal chance** of being selected.                   | • Using a **random number generator** to select students in a class to complete a task.                                                                             |
| **Stratified Sampling**   | • Smaller groups or **strata** within the sample are represented **proportionally** to the population.                                                 | • Finding out a favourite soap opera from different age **categories** of people in a year group.                                                                  |
| **Systematic Sampling**   | • Every member in the population is given a **number**. After the first member is chosen at random, the remaining members are chosen from an interval. | • A **list of people** with their first names in alphabetical order are numbered. The 5th person is **chosen randomly**, followed by every subsequent 8th person. |
| **Non Random Sampling**   | • **Convenience** sampling is used for **ease** of data collection. **Volunteers** usually collect data.                                               | • Asking people at a given location about how long their **commute** to work is.                                                                                    |
| **Capture Recapture**     | • **Collecting** a sample of data from **one location at different points in time**, **marking** the individuals to estimate a population size.         | • A sample of woodlice were **captured, marked and released**. Another sample of woodlice was captured 5 days later and the number of marked woodlice was counted. |

# **T-test**

---
## ✅ **What is a t-test?**

A **t-test** is a **statistical hypothesis test** used to determine whether there is a **significant difference between the means** of two groups. It's most useful when working with **small sample sizes** (typically < 30) and when the population standard deviation is **unknown**.

---
## 🎯 **Purpose / Use Case**

* To test if two sample means are **significantly different** from each other.
* Commonly used in:
  * **A/B testing**
  * **Medical trials**
  * **Behavioral studies**
  * **Comparing performance of models or methods**

---
## 📌 **Requirements / Assumptions**
1. **Data is continuous** (interval or ratio scale).
2. **Data follows a normal distribution** (for small samples).
3. **Samples are independent** (except in paired t-tests).
4. **Homogeneity of variance** (equal variances across groups — for unpaired tests).
5. **Random sampling** from the population.

---
## 🔢 **Types of t-tests & Formulas**
### 1. **One Sample t-test**
- Tests whether the mean of a single sample is significantly different from a known or hypothesized population mean $\mu_0$.
**Formula:**
$$
t = \frac{\bar{x} - \mu_0}{s / \sqrt{n}}
$$

Where:
* $\bar{x}$ = sample mean
* $\mu_0$ = population mean
* $s$ = sample standard deviation
* $n$ = sample size

---
### 2. **Two Sample t-test (Independent t-test)**
- Tests whether two **independent** sample means are significantly different.
**Formula:**

$$
t = \frac{\bar{x}_1 - \bar{x}_2}{\sqrt{s_p^2 \left( \frac{1}{n_1} + \frac{1}{n_2} \right)}}
$$

Where:
* $s_p^2$ = pooled variance
$$
s_p^2 = \frac{(n_1 - 1)s_1^2 + (n_2 - 1)s_2^2}{n_1 + n_2 - 2}
$$

---
### 3. **Paired Sample t-test**
- Used when the samples are **dependent** (e.g., before/after measurements on the same subjects).
**Formula:**

$$
t = \frac{\bar{d}}{s_d / \sqrt{n}}
$$

Where:
* $\bar{d}$ = mean of the differences
* $s_d$ = standard deviation of the differences
* $n$ = number of pairs
---
## 🧠 **Interpreting the Result**

* **p-value < 0.05**: Reject the null hypothesis ⇒ Significant difference exists.
* **p-value ≥ 0.05**: Fail to reject the null ⇒ No significant difference.

### ✅ One Sample t-test

````python
from scipy import stats

# Sample data
data = [2.3, 2.5, 2.8, 3.1, 2.9]
pop_mean = 3.0

t_stat, p_value = stats.ttest_1samp(data, pop_mean)

print(f"T-statistic: {t_stat}, P-value: {p_value}")
````

### ✅ Two Sample t-test (Independent)

````python
from scipy import stats

group1 = [2.9, 3.0, 3.2, 3.1, 2.8]
group2 = [3.5, 3.6, 3.4, 3.7, 3.8]

t_stat, p_value = stats.ttest_ind(group1, group2, equal_var=True)

print(f"T-statistic: {t_stat}, P-value: {p_value}")
````

### ✅ Paired Sample t-test

````python
from scipy import stats

before = [5.1, 5.3, 5.2, 5.0, 5.1]
after = [5.5, 5.6, 5.3, 5.4, 5.5]

t_stat, p_value = stats.ttest_rel(before, after)

print(f"T-statistic: {t_stat}, P-value: {p_value}")
````

# **Z-Test**
---

## ✅ **What is a Z-test?**

A **Z-test** is a **statistical hypothesis test** used to determine whether there is a **significant difference** between sample and population means (or between two means), **when the population standard deviation is known** and/or the sample size is large (typically $n > 30$).

---
## 🎯 **Use / Purpose**

* To compare a sample mean to a population mean.
* To compare two sample means.
* Common in:
  * **Large-scale A/B testing**
  * **Quality control**
  * **Business decisions**
  * **Comparing test scores, revenue, etc.**

---
## 📌 **Requirements / Assumptions**
1. **Data should be continuous.**
2. **Normal distribution** (especially for small samples).
3. **Known population standard deviation**.
4. **Sample size $n \geq 30$** for CLT to hold (Central Limit Theorem).
5. **Random and independent samples**.

---
## 🔢 **Formulas**
### 1. **One-Sample Z-test**
- Tests whether the sample mean $\bar{x}$ is significantly different from the population mean $\mu$.
- 
$$
Z = \frac{\bar{x} - \mu}{\sigma / \sqrt{n}}
$$

Where:
* $\bar{x}$: sample mean
* $\mu$: population mean
* $\sigma$: population standard deviation
* $n$: sample size

---
### 2. **Two-Sample Z-test**
- Used to compare two sample means $\bar{x}_1$ and $\bar{x}_2$, assuming known variances.

$$
Z = \frac{\bar{x}_1 - \bar{x}_2}{\sqrt{\frac{\sigma_1^2}{n_1} + \frac{\sigma_2^2}{n_2}}}
$$

---

## 🧠 Summary

| Test Type  | Use When                                  |
| ---------- | ----------------------------------------- |
| **Z-test** | Large $n$, known population std deviation |
| **t-test** | Small $n$, unknown population std         |

---

## 🔚 Summary Table

| Test Type              | When to Use                  | Formula Key Element     |   |      |
| ---------------------- | ---------------------------- | ----------------------- | - | ---- |
| One-Sample Z-test      | One group vs population mean | $\bar{x} - \mu$         |   |      |
| Two-Sample Z-test      | Two sample means comparison  | $\bar{x}_1 - \bar{x}_2$ |   |      |
| Z-test for Proportions | Two proportions comparison   | $p_1 - p_2$             |   |      |
| One-tailed Test        | Directional claim (>, <)     | $p = 1 - norm.cdf(z)$   |   |      |
| Two-tailed Test        | Non-directional (≠)          | ( p = 2(1 - norm.cdf(   | z | )) ) |

---

## 🧪 **Python Implementation of Z-test**
### 🧮 One-Sample Z-test

````python
import numpy as np
from scipy.stats import norm

# Sample data
data = [72, 74, 71, 69, 78, 76, 73, 75, 77, 74]
sample_mean = np.mean(data)
pop_mean = 70
pop_std = 5  # population std deviation (known)
n = len(data)

# Z-test
z_score = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))
p_value = 2 * (1 - norm.cdf(abs(z_score)))

print(f"Z-score: {z_score:.3f}")
print(f"P-value: {p_value:.3f}")
````

### 🧮 Two-Sample Z-test

````python
import numpy as np
from scipy.stats import norm

# Sample 1
x1 = [45, 49, 47, 50, 48, 46]
mean1 = np.mean(x1)
std1 = 4  # Assume known or from large n
n1 = len(x1)

# Sample 2
x2 = [42, 44, 41, 43, 45]
mean2 = np.mean(x2)
std2 = 3
n2 = len(x2)

# Z-test for means
z = (mean1 - mean2) / np.sqrt((std1**2 / n1) + (std2**2 / n2))
p = 2 * (1 - norm.cdf(abs(z)))  # Two-tailed

print(f"Z-score: {z:.3f}")
print(f"P-value (two-tailed): {p:.4f}")
````

### 🧮 Z-test for Proportions
- Used when comparing **proportions (percentages)** from two samples.

---
### 📌 Formula:
$$
Z = \frac{p_1 - p_2}{\sqrt{p(1 - p)\left(\frac{1}{n_1} + \frac{1}{n_2}\right)}}
\quad \text{where} \quad p = \frac{x_1 + x_2}{n_1 + n_2}
$$

* $x_1, x_2$: number of successes
* $n_1, n_2$: sample sizes
* $p_1 = x_1/n_1$, $p_2 = x_2/n_2$
---

````python
# Sample A
x1, n1 = 56, 100  # 56% success rate
# Sample B
x2, n2 = 48, 100

p1 = x1 / n1
p2 = x2 / n2
p_combined = (x1 + x2) / (n1 + n2)

# Z-test for proportions
z = (p1 - p2) / np.sqrt(p_combined * (1 - p_combined) * (1/n1 + 1/n2))
p = 2 * (1 - norm.cdf(abs(z)))

print(f"Z-score: {z:.3f}")
print(f"P-value (two-tailed): {p:.4f}")
````

# **F-Test**

---
## ✅ **What is an F-test?**
- An **F-test** is used to **compare the variances** (spread) of two populations to check if they are significantly different. It can also be used to compare **multiple means (ANOVA)** or test **model significance in regression**.

---
## 🎯 **Purpose of the F-test**

1. **Compare two population variances**
2. **Compare multiple group means (ANOVA)**
3. **Check significance in regression models**
4. **Test if a simpler model fits as well as a complex one**

---
## 📌 **Requirements / Assumptions**

* Samples are **independent**
* **Normally distributed** populations
* **Ratio of variances** is being tested
* Used when the data is **interval/ratio scale**

---
## 🔢 **Formula (Two-sample F-test)**

$$
F = \frac{s_1^2}{s_2^2}
$$

Where:

* $s_1^2$: variance of sample 1
* $s_2^2$: variance of sample 2
* $F$: follows an F-distribution with $df_1 = n_1 - 1$, $df_2 = n_2 - 1$

If $F \gg 1$ or $F \ll 1$, variances are significantly different.

---
## ✅ One-Tailed vs Two-Tailed F-test

* **F-tests are typically one-tailed**, since variance is always positive.
* **Two-tailed version** is only used when checking if variances are **not equal** (≠).
---
## 📚 Summary Table

| Use Case                | Test Name     | Library Function          |
| ----------------------- | ------------- | ------------------------- |
| Compare two variances   | F-test        | `scipy.stats.f`           |
| Compare 3+ group means  | ANOVA         | `scipy.stats.f_oneway()`  |
| Regression significance | F-test in OLS | `statsmodels` / `sklearn` |
---

## 🧪 Python Implementation (Two-sample F-test)

````python
#import
import numpy as np
from scipy.stats import f


alpha = 0.05

#Data
n1 = 10
n2 = 13
x1 = 15    #Sample mean
x2 = 14    #sample mean
ssd1 = 90
ssd2 = 108

#calculater variances
var_lays = ssd1 / (n1-1)
var_kur = ssd2 / (n2-1)
print(f"Variance Lays : {var_lays}")
print(f"Variance Kurkure : {var_kur}")

#calculate f_stats
f_statistic = max(var_lays,var_kur) / min(var_lays, var_kur)
print(f"F Statistic : {f_statistic}")

#degree of freedom
df1 = n1 - 1
df2 = n2 - 1

# #Critical value - do we need it in coding?   no!!!
# f_critical_value = f.ppf(1-alpha/2, df1,df2)
# print(f_critical_value)

p_value = f.cdf(f_statistic, df1, df2)      

if p_value < alpha:
    print("Reject Null Hypo: Variances are significantly diff than each other ")
else:
    print("Accept Null Hypo: Variances are significantly closer to each other")
````

# **2-way ANOVA**

## ✅ What is **Two-Way ANOVA**?

Two-way ANOVA tests the effect of **two independent categorical variables** on a **continuous dependent variable**, and also whether **there’s interaction** between those two variables.

---
## 📌 Use Case Example
> Imagine you're testing:
* **Factor A**: Type of fertilizer (F1, F2, F3)
* **Factor B**: Type of soil (Sandy, Clay, Loam)
* **Dependent Variable**: Plant growth

> A two-way ANOVA would tell you:

1. If **fertilizer type** affects growth
2. If **soil type** affects growth
3. If there’s an **interaction** between fertilizer and soil type

---
## ✅ Assumptions
1. Independent observations
2. Normality of the dependent variable
3. Equal variances (homoscedasticity)
4. Dependent variable is **continuous**, and factors are **categorical**
---
## ✅ Summary

| Test Component           | Tests For                         |
| ------------------------ | --------------------------------- |
| Main Effect A (Factor 1) | Does Factor A affect the outcome? |
| Main Effect B (Factor 2) | Does Factor B affect the outcome? |
| Interaction Effect (A×B) | Does Factor A interact with B?    |

---

## 🧪 Python Implementation (with `statsmodels`)

````python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Sample data
data = {
    'Fertilizer': ['F1', 'F1', 'F2', 'F2', 'F3', 'F3'] * 3,
    'Soil': ['Clay'] * 6 + ['Sandy'] * 6 + ['Loam'] * 6,
    'Growth': [12, 13, 14, 15, 16, 15, 10, 11, 13, 14, 15, 13, 18, 17, 19, 18, 20, 19]
}

df = pd.DataFrame(data)

# OLS model with interaction
model = ols('Growth ~ C(Fertilizer) + C(Soil) + C(Fertilizer):C(Soil)', data=df).fit()

# Two-way ANOVA table
anova_table = sm.stats.anova_lm(model, typ=2)
print(anova_table)
````

* **Main Effects**:

  * `C(Fertilizer)`: Effect of Fertilizer on Growth
  * `C(Soil)`: Effect of Soil on Growth

* **Interaction**:

  * `C(Fertilizer):C(Soil)`: Checks if the effect of Fertilizer changes depending on Soil type

> ✅ **If p-value < 0.05**, the factor or interaction has a significant effect.

## ✅ Visualization
This visual helps spot interactions: **non-parallel lines** suggest an interaction.

````python
import matplotlib.pyplot as plt
import seaborn as sns

sns.pointplot(x='Fertilizer', y='Growth', hue='Soil', data=df, dodge=True, markers='o', capsize=0.1)
plt.title("Interaction Plot: Fertilizer vs Soil on Growth")
plt.grid(True)
plt.show()
````

# **Chi-2 Test**

## ✅ What is the **Chi-Square Test**?

The **Chi-Square (χ²) Test** is a statistical method used to compare **observed frequencies** with **expected frequencies** to determine if there is a significant **association** between categorical variables.

---
## 📌 Types of Chi-Square Tests

| Test Type                           | Use Case                                                 |
| ----------------------------------- | -------------------------------------------------------- |
| **Chi-Square Goodness of Fit**      | Test if a sample matches a population distribution       |
| **Chi-Square Test of Independence** | Test if two categorical variables are related            |
| **Chi-Square Test of Homogeneity**  | Test if different populations have the same distribution |

---
## 🧠 Summary Table

| Element         | Meaning                                       |
| --------------- | --------------------------------------------- |
| χ² statistic    | Measures difference between observed/expected |
| df              | Degrees of freedom: (rows-1)(cols-1)          |
| p-value         | Significance level result                     |
| Expected Values | Computed under assumption of independence     |

---

## 🧪 Assumption Testing

* ✅ Use **`chi2_contingency()`** for independence
* ✅ Use **`chisquare()`** for goodness of fit

---

## 🧪 1. Chi-Square **Test of Independence**

Used to test **association between two categorical variables**.

📊 Example Scenario

|       | Likes Tea | Likes Coffee | Total |
| ----- | --------- | ------------ | ----- |
| Men   | 30        | 20           | 50    |
| Women | 25        | 25           | 50    |
| Total | 55        | 45           | 100   |

> Are **gender** and **beverage preference** independent?

---

### ✅ Assumptions

* Variables are categorical
* Observations are independent
* Expected frequencies ≥ 5 for most cells

---

### 🔢 Python Implementation

##### ✅ Interpretation:

* If **p-value < 0.05**, reject the null hypothesis — there's a **significant association** between gender and preference.
* If **p-value ≥ 0.05**, fail to reject — **no significant association**.

````python
import pandas as pd
from scipy.stats import chi2_contingency

# Create contingency table
data = [[30, 20],  # Men
        [25, 25]]  # Women

table = pd.DataFrame(data, columns=["Tea", "Coffee"], index=["Men", "Women"])

# Perform Chi-Square Test
chi2, p, dof, expected = chi2_contingency(table)

print("Chi-Square Statistic:", chi2)
print("Degrees of Freedom:", dof)
print("P-Value:", p)
print("Expected Frequencies:\n", expected)
````

---
## 🧪 2. Chi-Square **Goodness of Fit Test**
- Used to compare **observed distribution** to **expected distribution**.

### 📊 Example:
Observed colors in a candy pack:
* Red: 22, Blue: 18, Green: 20
Expected proportions: equal (i.e., 20 each if total is 60)

### 🔢 Python Code

````python
from scipy.stats import chisquare

observed = [22, 18, 20]
expected = [20, 20, 20]

chi2_stat, p_val = chisquare(f_obs=observed, f_exp=expected)

print("Chi-Square Statistic:", chi2_stat)
print("P-Value:", p_val)
````

## 📈 Visual Representation

````python
import matplotlib.pyplot as plt
import numpy as np

labels = ['Tea', 'Coffee']
observed = [30 + 25, 20 + 25]
expected = [55, 45]

x = np.arange(len(labels))
width = 0.35

fig, ax = plt.subplots()
ax.bar(x - width/2, observed, width, label='Observed')
ax.bar(x + width/2, expected, width, label='Expected')
ax.set_ylabel('Counts')
ax.set_title('Chi-Square Test: Observed vs Expected')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()
plt.show()
````

## 🧪 2. Chi-Square **Goodness of Fit Test** own code

````python
from scipy.stats import chi2

observed = [23,16,14,19,28]
expected = [20, 20, 20, 20, 20]

chi2_statistic = 0
for observed_val, expected_val in zip(observed,expected):
    chi2_statistic += ((observed_val - expected_val) ** 2) / expected_val
print("chi2_statistic : ",chi2_statistic)

df=4
alpha=0.01
print("alpha :", alpha)

#critical 
critical_value = chi2.ppf(1 - alpha,df)
print("critical_value :",critical_value)

p_value = 1 - chi2.cdf(chi2_statistic,df)
print(f"p_value : {p_value}")


if p_value < alpha:
    print("Reject H0 Hyp : there is sig diff in ob and ex")
else:
    print("Accept H0 Hyp: there is no sig diff in ob and ex")
````

````python
import numpy as np
import pandas as pd
from scipy import stats
from sklearn.feature_selection import chi2, f_classif
from sklearn.model_selection import train_test_split

# Sample dataset
np.random.seed(42)
data = {
    'Feature1': np.random.normal(50, 10, 100),  # Continuous variable
    'Feature2': np.random.normal(55, 15, 100),  # Continuous variable
    'Feature3': np.random.randint(0, 2, 100),   # Binary categorical variable
    'Target': np.random.randint(0, 2, 100)      # Binary target variable
}

df = pd.DataFrame(data)

# **T-test: Checking if Feature1 & Feature2 have different means**
t_stat, p_value = stats.ttest_ind(df['Feature1'], df['Feature2'])
print(f"T-test: t-stat={t_stat}, p-value={p_value}")

# **Z-test: Assuming large sample, compare means**
z_stat = (df['Feature1'].mean() - df['Feature2'].mean()) / np.sqrt(df['Feature1'].var()/len(df) + df['Feature2'].var()/len(df))
print(f"Z-test: z-stat={z_stat}")

# **F-test (ANOVA)**
f_stat, p_value = f_classif(df[['Feature1', 'Feature2']], df['Target'])
print(f"F-test (ANOVA) p-values: {p_value}")

# **Chi-Square Test for Categorical Data**
chi2_stat, p_value = chi2(df[['Feature3']], df['Target'])
print(f"Chi-Square Test: chi2-stat={chi2_stat}, p-value={p_value}")
````

````python
import numpy as np
import pandas as pd
from scipy import stats
from sklearn.feature_selection import chi2, f_classif
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# **1. Generate a Sample Dataset**
np.random.seed(42)
data = {
    'Age': np.random.randint(20, 60, 200),  # Continuous variable
    'BP': np.random.normal(120, 15, 200),   # Continuous variable
    'Cholesterol': np.random.normal(200, 50, 200),  # Continuous variable
    'Diabetes': np.random.randint(0, 2, 200),  # Binary categorical variable
    'Exercise': np.random.randint(0, 2, 200),  # Binary categorical variable
    'HeartDisease': np.random.randint(0, 2, 200)  # Binary Target Variable
}

df = pd.DataFrame(data)

# **2. Split into Features & Target**
X = df.drop(columns=['HeartDisease'])
y = df['HeartDisease']

# **3. Apply Statistical Tests for Feature Selection**

# **T-Test (Comparing means for Binary Features)**
t_stat, p_ttest = stats.ttest_ind(df['Age'][y==0], df['Age'][y==1])
print(f"T-test (Age vs HeartDisease): p-value = {p_ttest}")

# **F-test (ANOVA) for Continuous Features**
f_stat, p_anova = f_classif(X[['Age', 'BP', 'Cholesterol']], y)
print(f"F-test (ANOVA) p-values: {p_anova}")

# **Chi-Square Test for Categorical Features**
chi2_stat, p_chi2 = chi2(X[['Diabetes', 'Exercise']], y)
print(f"Chi-Square Test p-values: {p_chi2}")

# **Z-test for Large Samples (Comparing Means)**
z_stat = (df['BP'].mean() - df['Cholesterol'].mean()) / np.sqrt(df['BP'].var()/len(df) + df['Cholesterol'].var()/len(df))
print(f"Z-test (BP vs Cholesterol): z-stat = {z_stat}")

# **4. Select Significant Features (p < 0.05)**
selected_features = ['BP', 'Cholesterol', 'Diabetes']  # Based on p-values
X_selected = X[selected_features]

# **5. Split Data into Train and Test Sets**
X_train, X_test, y_train, y_test = train_test_split(X_selected, y, test_size=0.2, random_state=42)

# **6. Standardize Data (Important for Logistic Regression)**
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# **7. Train a Logistic Regression Model**
model = LogisticRegression()
model.fit(X_train, y_train)

# **8. Make Predictions and Evaluate Accuracy**
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Model Accuracy with Selected Features: {accuracy:.2f}")
````
