```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Load dataset (Example: Replace with your actual dataset)
np.random.seed(42)
data = np.random.normal(loc=50, scale=15, size=500)  # Normal distribution

# Convert to DataFrame
df = pd.DataFrame(data, columns=["Values"])

# Compute measures of central tendency
mean = np.mean(data)
median = np.median(data)

# Fix for mode: Find the bin with the highest count in histogram
hist_values, bin_edges = np.histogram(data, bins=30)
mode_bin_index = np.argmax(hist_values)  # Index of the bin with max count
mode = (bin_edges[mode_bin_index] + bin_edges[mode_bin_index + 1]) / 2  # Approximate mode

# Plot the distribution
plt.figure(figsize=(10, 5))
sns.histplot(data, bins=30, kde=True, color='blue', alpha=0.6)

# Mark mean, median, and mode
plt.axvline(mean, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {mean:.2f}')
plt.axvline(median, color='green', linestyle='dashed', linewidth=2, label=f'Median: {median:.2f}')
plt.axvline(mode, color='purple', linestyle='dashed', linewidth=2, label=f'Mode: {mode:.2f}')

plt.legend()
plt.title("Data Distribution with Measures of Central Tendency")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()

# Print inferences
print("### Inferences ###")
print(f"Mean: {mean:.2f}, Median: {median:.2f}, Mode: {mode:.2f}")
if mean > median:
    print("The distribution is right-skewed (positively skewed).")
elif mean < median:
    print("The distribution is left-skewed (negatively skewed).")
else:
    print("The distribution is symmetric (approximately normal).")

```
