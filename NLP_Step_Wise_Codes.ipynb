# Step 4 : **Stemming**

```python
import nltk

from nltk.tokenize import word_tokenize
nltk.download("punkt_tab")

from nltk.stem import PorterStemmer, LancasterStemmer, SnowballStemmer

import time
```

```python
text = "The various organizations were traditionally studying complex strategies for connecting diverse communities. Their researchers, known for analytical abilities and innovative thinking, were running multiple analyses daily. They hoped their findings would lead to positive changes and a beautiful revival of community spirit. It's generally considered a challenging task, but their commitment was unwavering. They believed in making connections and helping children, aiming for a fairly universal understanding. The university administration was cautiously observing the electricity consumption during these studies."
```

```python
porter_stemmer = PorterStemmer()
lancaster_stemmer = LancasterStemmer()
snowball_stemmer = SnowballStemmer(language = "english")
```

```python
word_tokens = word_tokenize(text)
print(word_tokens)
```

```python
start_time_porterstemmer = time.time() # time start
stemmed_tokens_porterstemmer = [porter_stemmer.stem(word) for word in word_tokens] # stemming
end_time_porterstemmer = time.time() # time stop

start_time_lancasterstemmer = time.time() # time start
stemmed_tokens_lancasterstemmer = [lancaster_stemmer.stem(word) for word in word_tokens] # stemming
end_time_lancasterstemmer = time.time() # time stop

start_time_snowballstemmer = time.time() # time start
stemmed_tokens_snowballstemmer = [snowball_stemmer.stem(word) for word in word_tokens] # stemming
end_time_snowballstemmer = time.time() # time stop
```

```python
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.expand_frame_repr', False)

# Create transposed DataFrame
df = pd.DataFrame({ "Original Word": [word for word in word_tokens],
                   "Porter Stemmer": [word for word in stemmed_tokens_porterstemmer],
                    "Lancaster Stemmer": [word for word in stemmed_tokens_lancasterstemmer],
                    "Snowball Stemmer": [word for word in stemmed_tokens_snowballstemmer]})
df.T
```

```python
# Show timing results
print("\n⏱ Execution Time (in milliseconds):")
print("Time taken by Porter Stemmer in ms    :", (end_time_porterstemmer - start_time_porterstemmer)*1000)
print("Time taken by Lancaster Stemmer in ms :", (end_time_lancasterstemmer - start_time_lancasterstemmer)*1000)
print("Time taken by Snowball Stemmer in ms  :", (end_time_snowballstemmer - start_time_snowballstemmer)*1000)
```

# Step 5 : **Lemmatization**

```python
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
nltk.download("punkt_tab")

from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
nltk.download("wordnet")
nltk.download("omw-1.4") # for lemmetizing words like better, it is recommended to do lemmetization with POS tagging

import time
```

```python
text = "The various organizations were traditionally studying complex strategies for connecting diverse communities. Their researchers, known for analytical abilities and innovative thinking, were running multiple analyses daily. They hoped their findings would lead to positive changes and a beautiful revival of community spirit. It's generally considered a challenging task, but their commitment was unwavering. They believed in making connections and helping children, aiming for a fairly universal understanding. The university administration was cautiously observing the electricity consumption during these studies."
```

```python
word_tokens = word_tokenize(text)
```

```python
lemmatizer = WordNetLemmatizer()

start_time_lemmatization = time.time() # time start
lemmatized_words = [lemmatizer.lemmatize(word) for word in word_tokens] # Lemmatization
end_time_lemmatization = time.time() # time stop

elapsed_time_lemmatization = (end_time_lemmatization - start_time_lemmatization)*1000
```

```python
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.expand_frame_repr', False)

# Create transposed DataFrame
df = pd.DataFrame({"Original Word": [word for word in word_tokens],
                   "Lemmatized word": [word for word in lemmatized_words]})
df.T
```

```python
print("\n⏱ Execution Time (in milliseconds):")
print("Time taken by Lemmatization in ms :", elapsed_time_lemmatization)
```

## **Lemmatization using POS Tags**

```python
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
nltk.download("punkt_tab")

from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
nltk.download("wordnet")
nltk.download("omw-1.4") # for lemmetizing words like better, it is recommended to do lemmetization with POS tagging

import time

from nltk import pos_tag
nltk.download("averaged_perceptron_tagger_eng")
```

```python
text = "The various organizations were traditionally studying complex strategies for connecting diverse communities. Their researchers, known for analytical abilities and innovative thinking, were running multiple analyses daily. They hoped their findings would lead to positive changes and a beautiful revival of community spirit. It's generally considered a challenging task, but their commitment was unwavering. They believed in making connections and helping children, aiming for a fairly universal understanding. The university administration was cautiously observing the electricity consumption during these studies."
```

```python
# Helper function
def get_wordnet_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN # Default

pos_tags = pos_tag(word_tokens)
print(pos_tags)
```

```python
lemmatized_words = [lemmatizer.lemmatize(word, get_wordnet_pos(pos)) for word, pos in pos_tags]
print("\nWith Pos tagging : ")
print("Original words : ", word_tokens)
print("Lemmatized words : ", lemmatized_words)
```

```python
# Customizing the conversion
sub_text = ["good", "better", "best","worse"]
adj_map = {"better" : "good", "best" : "good", "worse" : "bad", "worst" : "bad"}
lemmatized = [adj_map.get(w, lemmatizer.lemmatize(w, pos=wordnet.ADJ)) for w in sub_text]
print(lemmatized)
```

# **WordCloud**

```python
# !pip install wordcloud
from wordcloud import WordCloud, STOPWORDS

import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
nltk.download('stopwords')
nltk.download('punkt')

import matplotlib.pyplot as plt
```

```python
positive_reviews = """
The flight was incredibly smooth and the crew was very courteous. I loved the in-flight entertainment and the meals were surprisingly delicious.
Excellent service from the check-in to the landing. Seats were comfortable and the staff was always smiling.
I had a wonderful experience flying with this airline. On-time departure and arrival, and clean cabins.
Top-notch service and friendly attendants. Will definitely fly again!
The legroom in economy was better than expected. The flight attendants were attentive and kind.
Booking was easy, check-in was fast, and the plane was spotless. Great experience overall.
Very professional staff and the captain kept us informed throughout the flight.
Loved the new aircraft and the onboard Wi-Fi worked flawlessly.
Highly satisfied with their service. Luggage arrived on time and undamaged.
Seats reclined well and there were USB ports to charge devices. Great airline!
"""
```

```python
tokens = word_tokenize(positive_reviews)
cleaned_text = " ".join([word for word in tokens if word.lower() not in stopwords.words("english")])
```

```python
wordcloud = WordCloud(stopwords = STOPWORDS,
                      width = 800,
                      height = 400,
                      background_color = "Black",
                      colormap = 'rainbow',
                      min_font_size = 2
                      ).generate(cleaned_text)

plt.figure(figsize = (10,5))
plt.imshow(wordcloud)
plt.axis("off")
plt.title('Positive Reviews Word Cloud')
plt.show()
```

```python
negative_reviews = """
The flight was delayed for over 3 hours and no proper communication from the staff.
Worst experience ever. Lost my luggage and customer service was rude and unhelpful.
Seats were cramped and dirty. The cabin was smelly and hot.
Unfriendly crew and no water provided during a 5-hour flight.
In-flight entertainment system was broken and they didn’t offer any compensation.
Food was cold and tasteless. Not worth the money.
I was charged extra for carry-on baggage at the gate which wasn’t mentioned before.
Check-in process was chaotic and disorganized.
Plane was old and the overhead bins were barely functional.
The flight attendants were dismissive and ignored requests. Never flying with them again.
"""

# Generate word cloud for negative reviews
wordcloud_neg = WordCloud(width=800, height=400, background_color='black', colormap='Reds').generate(negative_reviews)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud_neg, interpolation='bilinear')
plt.axis('off')
plt.title('Negative Reviews Word Cloud')
plt.show()
```

# **Count Vectorizer**
- CountVectorizer is a tool in scikit-learn that coverts a collection of text documents to a matrix of token counts.
- It is a form of BOW (Bag-of-words) model which simply counts how often each word occurs in a document.

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
text = ["I love machine learning",
        "Machine learning is fun",
        "I enjoy learning"]
```

```python
count_vectorizer = CountVectorizer(max_features = 100)
data_features = count_vectorizer.fit_transform(text)
print(data_features)
```

```python
data_features_to_array = data_features.toarray()

print("Vocabulary : ",count_vectorizer.get_feature_names_out(),"\n")
print("Data Features : \n",data_features_to_array)
```

In above the array each column is representing the corresponding word in "Vocabulary" respectively. and 0 and 1 tells the presence and absence of that vocabulary
